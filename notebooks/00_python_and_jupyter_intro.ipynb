## Part 2: What Are We Studying?

### The Particle Physics Zoo

Everything around us is made of tiny particles. The **Standard Model** of particle physics describes all known fundamental particles:

- **Quarks**: Building blocks of protons and neutrons (6 types)
- **Leptons**: Includes electrons and neutrinos (6 types)
- **Force carriers**: Photons, gluons, W/Z bosons
- **The Higgs boson**: Gives particles mass

### Meet the Neutrino: The Ghost Particle

Neutrinos are fascinating because they:
- Have almost no mass (a million times lighter than an electron!)
- Don't have electric charge
- Barely interact with matter (trillions pass through you every second!)
- Come in three "flavors": electron ($\nu_{e}$), muon ($\nu_{μ}$), and tau ($\nu_{\tau}$)

### The Mystery: Neutrino Oscillations

Here's the strange part: **neutrinos can change flavor as they travel!**

A muon neutrino (νμ) created in a beam might become an electron neutrino (νₑ) after traveling hundreds of kilometers. This is called **neutrino oscillation**, and it only happens because neutrinos have mass.

**Why does this matter?**
- It proves the Standard Model is incomplete (originally predicted massless neutrinos)
- It might explain why our universe has more matter than antimatter
- Understanding it could reveal new physics!

### How Do We Detect Them?

Since neutrinos barely interact, we need **huge detectors** and **lots of neutrinos**:

1. **NOvA** (NuMI Off-Axis νₑ Appearance): Two detectors 810 km apart
   - Near detector at Fermilab (Illinois)
   - Far detector in Minnesota
   
2. **DUNE** (Deep Underground Neutrino Experiment): Future detector in South Dakota

The neutrino beam is created by:
1. Smashing protons into a target
2. Producing pions (π)
3. Pions decay into muons and neutrinos

**The key idea**: Compare what we see at the **near detector** (no oscillation yet) vs. the **far detector** (oscillations have occurred). The difference reveals oscillation!

GitHub is like "Google Docs for code" - it lets you save your work, track changes, and collaborate.


### Key Terms:
- **Repository (repo)**: A project folder tracked by git
- **Commit**: A saved snapshot of your work
- **Branch**: A separate line of work, off to the side of the main project
- **Push**: Upload your commits to GitHub
- **Pull**: Download changes from GitHub
- **Pull Request (PR)**: A request to merge your branch's changes into `main`, opened so others can review before it happens

**Pro tip**: Commit often! It's like saving a video game - you can always go back to earlier versions.

### Why Branches?

Everyone in MESH is editing notebooks in the *same* repository at the same time. If everyone committed straight to `main`, your changes and someone else's could collide, or a half-finished edit could break the notebook for the whole group.

A **branch** solves this: it's your own copy of the project to work in. Nothing you do there affects `main`, or anyone else, until you explicitly ask for it to be merged.

```bash
# Create a branch and switch to it in one step
# Naming convention: yourname/short-description
git checkout -b alex/exercise-2-histogram

# ... edit, save, and commit as usual on this branch ...
git add notebook_name.ipynb
git commit -m "Complete exercise 2"

# Push the branch (the -u only needed the first time you push it)
git push -u origin alex/exercise-2-histogram
```

### Pull Requests: Asking to Merge Your Work

Once your branch is pushed, GitHub will offer to open a **Pull Request (PR)** - a page that shows exactly what changed, line by line, and lets an instructor (or teammate) review it before it becomes part of `main`.

1. On GitHub, click **Compare & pull request** for your branch (or open one from the "Pull requests" tab).
2. Give it a title describing what you did, e.g. "Complete exercise 2: energy histogram."
3. Open the PR. An instructor will review and either merge it or leave comments asking for changes.
4. Once merged, your branch's commits become part of `main` - and everyone else's clone will see them the next time they `git pull`.

This is the same workflow used by real physics software projects (including `nuosclab`, which this curriculum is built on) - nothing gets into `main` without a reviewed PR.

## Part 4: Your First Physics Plot - Making a Histogram

### What is a Histogram?

A **histogram** shows how often different values appear in your data. In particle physics, we use histograms to show:
- Energy distributions of particles
- Number of events detected
- And much more!

### Example: Rolling Dice

Let's start with something familiar. If we roll a die 1000 times, we expect each number (1-6) to appear about 1/6 of the time. Let's test it!

````python
# Simulate rolling a die 1000 times
rolls = np.random.randint(1, 7, size=1000)

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(rolls, bins=6, range=(0.5, 6.5), edgecolor="black", alpha=0.7)
plt.xlabel("Die Face")
plt.ylabel("Number of Rolls")
plt.title("Distribution of 1000 Die Rolls")
plt.xticks([1, 2, 3, 4, 5, 6])
plt.grid(True, alpha=0.3)
plt.show()

# Print statistics
print(f"Average roll: {np.mean(rolls):.2f} (expected: 3.5)")
print(f"Most common: {np.bincount(rolls)[1:].argmax() + 1}")
````

### Physics Example: Particle Energies

Now let's make a histogram that looks more like real physics data. Imagine we're detecting particles and measuring their energies:

````python
# Simulate detected particle energies (in GeV)
# Real data would come from a detector, but we'll generate fake data for now
np.random.seed(42)  # Makes results reproducible

# Most particles have energy around 2 GeV (a "peak")
energies = np.random.normal(loc=2.0, scale=0.5, size=500)

# Remove unphysical negative energies
energies = energies[energies > 0]

print(f"Detected {len(energies)} particles")
print(f"Energy range: {energies.min():.2f} to {energies.max():.2f} GeV")
````

````python
# Create a histogram of particle energies
plt.figure(figsize=(10, 6))

counts, bins, patches = plt.hist(
    energies, bins=30, edgecolor="black", alpha=0.7, color="steelblue"
)

plt.xlabel("Energy (GeV)", fontsize=12)
plt.ylabel("Number of Particles", fontsize=12)
plt.title("Particle Energy Distribution", fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3)

# Add a vertical line at the mean energy
mean_energy = np.mean(energies)
plt.axvline(
    mean_energy,
    color="red",
    linestyle="--",
    linewidth=2,
    label=f"Mean: {mean_energy:.2f} GeV",
)

plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

print("\nStatistics:")
print(f"  Mean energy: {mean_energy:.2f} GeV")
print(f"  Median energy: {np.median(energies):.2f} GeV")
print(f"  Std deviation: {np.std(energies):.2f} GeV")
````

### Understanding the Histogram

What does this histogram tell us?
- **X-axis**: Particle energy (in GeV, a unit of energy)
- **Y-axis**: How many particles had that energy
- **Peak**: Most particles have energy around 2 GeV
- **Spread**: Some particles have higher or lower energy (this is natural variation)

In real Belle II analyses, we'll see histograms showing:
- How many events pass a given selection
- How distributions **change** when we apply physics-motivated cuts

## Part 4.5: Real Data, pandas, and Cuts

So far we've made histograms from data we generated ourselves. Real experiments store their data in tables (rows and columns) - a **pandas DataFrame** is the standard Python tool for working with tables like this.

Let's load a small example dataset: travel times (in minutes) for students commuting to the physics department.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a shared spreadsheet
url = "https://docs.google.com/spreadsheets/d/1y7QJVaC3WmaBoi7BtqijTwLQbOqsn9MKLuecsq86Osw/export?format=csv"
data = pd.read_csv(url)

# Check the data
print(data.columns.tolist())
data.head()
````

````python
# Plot a histogram of travel times for all students
plt.figure(figsize=(8, 5))
plt.hist(data["travel_time(min)"], bins=20, edgecolor="black", alpha=0.7)
plt.title("Travel Time Distribution (All Students)")
plt.xlabel("Time (minutes)")
plt.ylabel("Number of Students")
plt.show()
````

### Cuts

**Cuts** are used to filter out data that isn't relevant to what you're studying, or that would distort your result. You define a cut based on the goal of your analysis, then keep only the rows that pass it.

In NOvA, cuts are applied all the time to isolate meaningful events from large amounts of raw data - for example, cutting on energy, particle identification (PID), or L/E ratio to select well-reconstructed neutrino candidates.

Here, let's say we only want students whose commute is **80 minutes or less** (e.g., we think anything longer is an outlier, or we're only interested in local commuters). We apply the cut by keeping only the rows where that condition is `True`:

````python
# Define the cut: keep only rows where travel_time is 80 minutes or less
cut = data["travel_time(min)"] <= 80
data_cut = data[cut]

print(f"Before cut: {len(data)} students")
print(f"After cut:  {len(data_cut)} students")

# Compare before vs. after, on the same axes
plt.figure(figsize=(8, 5))
plt.hist(
    data["travel_time(min)"], bins=20, alpha=0.5, label="Before cut", edgecolor="black"
)
plt.hist(
    data_cut["travel_time(min)"],
    bins=20,
    alpha=0.7,
    label="After cut (≤ 80 min)",
    edgecolor="black",
)
plt.title("Travel Time Distribution: Before vs. After Cut")
plt.xlabel("Time (minutes)")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
````
